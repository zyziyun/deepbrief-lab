## What you'll build by the end

A multi-agent system that takes a topic and produces a structured research brief with citations:

```
you: "State of WebGPU adoption in 2026"
      ↓
  Coordinator → decomposes into sub-questions
      ↓ (A2A)
  Researchers → run in parallel, call MCP tools (search + fetch)
      ↓
  Synthesizer → merges findings
      ↓
  Editor (HITL) → you approve before saving
      ↓
  brief.md with citations
```

## 1. Verify environment variables

````python
import os
from dotenv import load_dotenv

load_dotenv()

openai_ok = bool(os.getenv("OPENAI_API_KEY"))
tavily_ok = bool(os.getenv("TAVILY_API_KEY"))

print(f"OPENAI_API_KEY:  {'✅ set' if openai_ok else '❌ missing — required for notebook 01+'}")
print(f"TAVILY_API_KEY:  {'✅ set' if tavily_ok else '⚠️  missing — required for notebook 04+'}")

assert openai_ok, "Set OPENAI_API_KEY in .env before continuing."
````

## 2. Test the OpenAI connection

If this cell fails, it's almost always one of:
- Wrong key (typos, expired)
- Network issue (corporate proxy, firewall)
- Out of credits

````python
from openai import AsyncOpenAI

client = AsyncOpenAI()

resp = await client.chat.completions.create(
    model=os.getenv("OPENAI_MODEL", "gpt-4o-mini"),
    messages=[{"role": "user", "content": "Say hi to a DeepBrief student in 8 words or fewer."}],
)
print(resp.choices[0].message.content)
print(f"\nmodel: {resp.model}  •  tokens: {resp.usage.total_tokens}")
````

## 3. (Optional) Test Tavily — only needed from notebook 04 onward

````python
if tavily_ok:
    from tavily import TavilyClient
    tv = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))
    r = tv.search("What is the Model Context Protocol", max_results=2)
    for hit in r["results"][:2]:
        print("•", hit["title"])
        print("  ", hit["url"])
else:
    print("Skipped — set TAVILY_API_KEY before notebook 04.")
````
